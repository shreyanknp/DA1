```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
print(iris.head(10))

sample_means = []
for _ in range(100):  # Repeat sampling 100 times
    sample = iris.sample(n=30, random_state=np.random.randint(0, 10000))  # Random sample of 30 observations
    sample_mean = sample['sepal_length'].mean()  # Calculate mean sepal length
    sample_means.append(sample_mean)

plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Sample Means (Sepal Length)')
plt.xlabel('Mean Sepal Length')
plt.ylabel('Frequency')
plt.show()

step=int(1/0.2) #sampling 20% of dataset
index=np.arange(0, len(iris), step)
sampled_data=iris.iloc[index]
sampled_data.head(10)

plt.figure(figsize=(12, 6))

# Original dataset
plt.subplot(1, 2, 1)
sns.histplot(iris['sepal_length'], color='blue')
plt.title('Original Dataset (Sepal Length)')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

# Sampled dataset
plt.subplot(1, 2, 2)
sns.histplot(sampled_data['sepal_length'], color='orange')
plt.title('Systematic Sample (20%)')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

#plt.tight_layout()
plt.show()
```
