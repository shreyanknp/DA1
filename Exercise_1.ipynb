```python
#from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import scipy.stats as stats
import statistics


#path='/content/sample_data/california_housing_test.csv'
path='/content/drive/MyDrive/Datasets/StatewiseTestingDetails.csv'

df=pd.read_csv(path)
df.head(10)
total_samples=df.TotalSamples
mean=total_samples.mean()
median=total_samples.median()
mode=statistics.mode(total_samples)
print(f'Mean is {mean},', f'Median is {median},', f'Mode is {mode},')

standard_deviation=total_samples.std()
variance=total_samples.var()
range=max(total_samples)-min(total_samples)
print(f'Standard Deviation is {standard_deviation},', f'Variance is {variance},', f'Range is {range},')

plt.hist(x=[total_samples], bins=30, color='skyblue', edgecolor='black')
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1)
plt.axvline(x=median, color='green', linestyle='-.', linewidth=1)
plt.axvline(x=mode, color='yellow', linestyle='-.', linewidth=1)
plt.axvline(x=standard_deviation, color='orange', linestyle='-', linewidth=1)
plt.axvline(x=range, color='orange', linestyle='--')
plt.xlabel('Total Samples')
plt.ylabel('Frequency')
plt.title('Histogram of Total Samples')
plt.legend(['Mean', 'Median', 'Mode','Standard Deviation','Range'])
plt.show()
```
